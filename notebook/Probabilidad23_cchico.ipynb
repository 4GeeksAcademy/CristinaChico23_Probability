## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# Eventos / Sucesos (Casos Favorables): suma de los dos dados mayor a 7 o número par 
# Experimentos (Acción o procedimiento para definir el resultado): Lanzar 2 dados 1000 veces
# Espacio muestral (Casos Posibles): ?
#cuantas vees suma 2 dados == 4

import random

experimentos = 1000

def tirar_dado1():
    dado1 = random.randrange(1,7)
    return dado1   

def tirar_dado2():
    dado2 = random.randrange(1,7)
    return dado2   

caso_favorable = 0
for tirada in range(experimentos):
    dado1=tirar_dado1()
    dado2=tirar_dado2()
    sum = dado1+dado2
    if sum >7 or sum%2== 0:
        caso_favorable +=1
        
probabilidad_empirica = caso_favorable/experimentos
print(probabilidad_empirica)
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
# ball_box = {}

# Crea la cajas con las bolas
# for i in range(60):
#     if i < 10:
#         ball_box[i] = "White"
#     elif (i > 9) and (i < 30):
#         ball_box[i] = "Red"
#     else:
#         ball_box[i] = "Green"

# print(ball_box)
            
```

```python
import numpy as np

ball_box = {i: "White" if i < 10 else "Red" if i < 30 else "Green" for i in range(60)} # Caja de bolas original


ball_list = [] # Lista que acumula las bolas que se sacan de la ball_box en cada iteración ?¿

for ball in ball_box:
    if ball < 10:
        ball_list.append("White")
    elif ball <30:
        ball_list.append('Red')
    else:
        ball_list.append('Green')

# Definir variables de inicio: Conteo de bolas por color
white = ball_list.count('White')
red = ball_list.count('Red')
green = ball_list.count('Green')

simulations = 1000 # Número de experimentos

# Contadores para almacenar la suma del resultado.
counter_white_red = 0
counter_all_colors = 0

# Comprobar la probabilidad de sacar 3 y 2. 
# # ¿?¿? porque simulo 5 ¿?¿?¿? Simulo 5 extracciones(samples) aleatorias

for _ in range(simulations): 
    random_extract = np.random.choice(ball_list, 5, replace=False)  
    white = np.count_nonzero(random_extract=='White')
    red = np.count_nonzero(random_extract=='Red')
    green = np.count_nonzero(random_extract=='Green')

    if white == 3 and red == 2:  # Iterar la lista con la condición dada
        counter_white_red += 1

    if white == 5 or red == 5 or green == 5: # Comprobar la probabilidad de sacar la misma cantidad de bolas del mismo color. 
        counter_all_colors += 1
    
# Cálculo de la probabilidad por evento.
probability_red_white = counter_white_red / simulations
probability_all_colors = counter_all_colors / simulations
```

```python
# Resultado probabilidad 3 bolas blancas y 2 rojas:
f'Probabilidad de 3 blancas y 2 rojas: {probability_red_white}'
```

```python
# Resultado probabilidad todas las bolas del mismo color:
f'Probabilidad de todas las bolas del mismo color: {probability_all_colors}'
```
